# EDA  on  mobile price classification

The taken dataset contain the following features to classify the price of the Mobile.the below features are included in the dataset.

* battery_power : Total energy a battery can store in one time measured in mAh
* blue : Has bluetooth or not
* clock_speed : speed at which microprocessor executes instructions
* dual_sim : Has dual sim support or not
* fc : Front Camera mega pixels
* four_g : Has 4G or not
* int_memory : Internal Memory in Gigabytes
* m_dep : Mobile Depth in cm
* mobile_wt : Weight of mobile phone
* n_cores : Number of cores of processor
* price_range : price range
* mobile_wt : Weight of mobile phone
* pc : Primary Camera mega pixels
* px_height : Pixel Resolution Height
* px_width : Pixel Resolution Width
* ram : Random Access Memory in Megabytes
* sc_h : Screen Height of mobile in cm
* sc_w : Screen Width of mobile in cm
* talk_time : longest time that a single battery charge will last when you are
* touch_screen : Has touch screen or not
* wifi : Has wifi or not

# Feature selection using  Extra tree classifier algorithim

```python
#extracting top 10 featutes that will reponsible for price range classification
X=train.iloc[:,0:20]
Y=train['price_range']
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(X, Y)
scores=model.feature_importances_
feat_importances = pd.Series(scores, index=X.columns)
feat_importances=feat_importances.sort_values(ascending=False)[:10]
plt.figure(figsize=(20,10))
sns.barplot(x=feat_importances.index,y=feat_importances.values)
top_10_list=feat_importances.index
```

```python
top_10_list=feat_importances.index
top_10_list
```

# Info gain to select most relavant features

```python
#removing the features which are not corrrelated with target variable(having mutual_info=0)
from sklearn.feature_selection import mutual_info_classif
mutual_info=mutual_info_classif(X,Y)
mutual_data=pd.Series(mutual_info,index=X.columns)
mutual_data.sort_values(ascending=False)
```

*  we have  to remove sc_w feature,talkt_time feature,pc feature since the mutual information gain from this features(columns)==0

# Dropping the features that are not useful

```python
#dropping redundant columns from the data
data=train.drop(['wifi','touch_screen','three_g','talk_time','sc_w','dual_sim','four_g','int_memory','blue','n_cores','mobile_wt','m_dep','fc'],axis=1)
```

```python
#final data for the model building
data
```

# Splitting the data for training and testing

```python
# splitting data for traing the model and testing model
X=data.iloc[:,0:7]
Y=data['price_range']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.33, random_state = 0)
```

# Normalizing the data(Data Scaling)

```python
#Data normalizing for better perfomance of the model
from sklearn.preprocessing import MinMaxScaler
X_train=MinMaxScaler().fit_transform(X_train)
X_test=MinMaxScaler().fit_transform(X_test)
X_test
```

# Applying logistic regression for mobile_price Classification.

```python
#creating the model and applying  the model 
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
predicted_class= model.predict(X_test)
```

```python
predicted_class
```

```python
y_test=np.array(y_test)
y_test
```

# Performance measure 

```python
# measuring accuracy between Y_test and predicted price class
from sklearn.metrics import accuracy_score
acc1= accuracy_score(y_test,predicted_class)
acc1
```

**The Accuracy of the logistic regression is 95%**

#  Confusion matrix for logistic regresssion

```python
#Confusion matrix for price classification
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,predicted_class)
cm
```

```python
#applying Decision tree algorithim 
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
predicted_class= model.predict(X_test)
```
